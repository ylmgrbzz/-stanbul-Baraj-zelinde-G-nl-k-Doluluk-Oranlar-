```python
def barajgetir(tarih,barajisim):
  baraj=pd.read_csv("baraj.csv",encoding="latin1",skiprows=1,delimiter=";")
  baraj=baraj.set_index("Tarih")
  sonuc=baraj.loc[tarih][barajisim]
  sonuc=sonuc.replace(",",".") 
  sonuc=sonuc.replace("%","") 
  sonuc=float(sonuc) 
  return sonuc
```
